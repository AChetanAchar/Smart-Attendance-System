```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read dataset
data = pd.read_csv('Students_details.csv')

# 1. Total Students
print("The Total Number of Students in the Class are", len(data))

# 2. Boys vs Girls count
gender_counts = data['Gender'].value_counts()
print("The Number of Boys in the Class are", gender_counts.get('Male', 0))
print("The Number of Girls in the Class are", gender_counts.get('Female', 0))

# 3. Sports Interest
print("\nStudents Who are Interested in Sports Other Than Studies:\n")
sports_counts = data['Favorite_Sport'].value_counts()
print(sports_counts)

# Most popular sport
print("\nThe Most Popular Sport is:", sports_counts.idxmax(), "with", sports_counts.max(), "students")

# 4. Visualization - Sports Popularity
sports_counts.plot(kind='bar', title="Students' Interest in Different Sports", color='skyblue')
plt.xlabel("Sport")
plt.ylabel("Number of Students")
plt.show()

# 5. Low GPA but strong sports passion
low_gpa_sports = data[(data['Grade'] < 7.5) & (data['Hours_per_Week'] >= 8)]
print("\nStudents who got low GPA but are very passionate about sports:\n")
print(low_gpa_sports[['Student_ID', 'Gender', 'Grade', 'Favorite_Sport', 'Hours_per_Week']])
print("\nTotal such students:", len(low_gpa_sports))

# 6. Scatter Plot - GPA vs Sports Hours
plt.scatter(data['Hours_per_Week'], data['Grade'], c='blue', alpha=0.6)
plt.title("GPA vs Sports Hours")
plt.xlabel("Hours per Week spent on Sports")
plt.ylabel("Grade (GPA)")
plt.grid(True)
plt.show()

# 7. Line Graph - GPA Trend by Sports Hours
sorted_data = data.sort_values(by='Hours_per_Week')
plt.plot(sorted_data['Hours_per_Week'], sorted_data['Grade'], marker='o', linestyle='-', color='green')
plt.title("Trend: GPA vs Sports Hours")
plt.xlabel("Hours per Week spent on Sports")
plt.ylabel("Grade (GPA)")
plt.grid(True)
plt.show()

# 8. Bar Chart - Hours per Student
plt.figure(figsize=(12,5))
plt.bar(data['Student_ID'], data['Hours_per_Week'], color='orange')
plt.xticks(rotation=90)
plt.title("Sports Hours per Student")
plt.xlabel("Student ID")
plt.ylabel("Hours per Week")
plt.show()
```
